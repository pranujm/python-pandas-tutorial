# Load data using Pandas

```python
import pandas as pd

df = pd.read_csv('pokemon_data.csv') # load csv file 
print(df.head(3)) # print the first 3 rows of data
```

# Read data using Pandas

```python
print(df.columns) # read headers
print(df[['Name', 'Type 1', 'HP']]) # print each column
# print(df['Name'][0:5]) # print first 5 columns
```

```python
print(df.head(4)) # print first 4 rows
print(df.iloc[1])  # print everything from one of the rows

# print(df.iloc[1:4])  # print everything from rows 1 to 4

print(df.iloc[2, 1]) # read a specific location (R, C)
```

```python
# iterate through each row (in this case, by name)
for index, row in df.iterrows():
    print(index, row['Name']) 

# get row based on specific condition (in this case, all fire types)
df.loc[df['Type 1'] == 'Fire'] 
```

# Sorting & describing data

```python
df.describe() # describes dataset
```

```python
df.sort_values('Name', ascending = False) # sorts by name descending
```

```python
df.sort_values(['Type 1', 'HP'], ascending = [1, 0]) # sorts by type 1 ascending and HP descending
```

# Making changes to data

```python
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed'] # makes new column for base stat totals

df = df.drop(columns = ['Total']) # removes the specified column

# another way to add total
#colon means we are changing all rows, and stat_columns are the columns we are using in each of those rows
# sum means we are adding the numbers in those columns for each row, and axis = 1 means we are adding the numbers horizontally
stat_columns = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
df['Total'] = df[stat_columns].sum(axis = 1) 

df.head(5) # show first 5 rows of data
```

```python
# Remove the Total column and insert it after Speed

total_values = df.pop('Total')  # Remove and store the values
speed_index = df.columns.get_loc('Speed')  # Find Speed column position
df.insert(speed_index + 1, 'Total', total_values)  # Insert after Speed
df.head(5)
```

# Saving our data (export into desired format)

```python
df.to_csv('modified.csv') # creates new csv of modified data 
```

# Filtering data

```python
df.loc[df['Type 1'] == 'Dragon'] # filters out everything except for dragon types
```

```python
# filters out everything except for type 1 normal and type 2 flying types
# OR sign (|) means all pokemon with either typing in their respective columns
df.loc[(df['Type 1'] == 'Normal') | (df['Type 2'] == 'Flying')]
```

```python
# filters out everything except for normal/flying types
# AND sign (&) means all pokemon with both types in their respective columns, and HP stat is above 70
new_df = df.loc[(df['Type 1'] == 'Normal') & (df['Type 2'] == 'Flying') & (df['HP'] > 70)]

# drop method replaces old indexes with no modified indexes, and inplace = True means no need to assign to new variable
new_df.reset_index(drop = True, inplace = True) 
print(new_df)

new_df.to_csv('filtered.csv') # creates new filtered csv file
```

```python
df.loc[df['Name'].str.contains('Mega')] # filters out everything except for pokemon with 'Mega' in the name
```

```python
# NOT sign (~) means the reverse/opposite of the line, so it actually removes all Megas instead 

df.loc[~df['Name'].str.contains('Mega')] # filters out all pokemon with 'Mega' in the name
```

```python
import re # imports regular expressions (very powerful for filtering data)

# regex = True allows us to filter out all pokemon except for those with fire OR grass as their first types
# flags = re.I means ignore case, allowing us to get the result without having to capitalize the column names 

df.loc[df['Type 1'].str.contains('fire|grass', flags = re.I, regex = True)]
```

```python
# sorts for all pokemon from A - Z with 'pi' in its name
df.loc[df['Name'].str.contains('pi[a-z]*', flags = re.I, regex = True)]

# The '^' symbol allows us to get pokemon whos names START with 'pi'
df.loc[df['Name'].str.contains('^pi[a-z]*', flags = re.I, regex = True)]
```

# Aggregate statistics (Groupby)

```python
df = pd.read_csv('modified.csv')

# computing the average values of each pokemon grouped by their primary
df.groupby(['Type 1']).mean(numeric_only = True)
```

```python
# sort by which primary type has the highest average defense
df.groupby(['Type 1']).mean(numeric_only = True).sort_values('Defense', ascending = False) 
```

```python
# compute the total stat values for all pokemon of their primary types
df.groupby(['Type 1']).sum(numeric_only = True)
```

```python
# counts how many pokemon have each type as their primary type

df['count'] = 1
df.groupby(['Type 1']).count()['count']
```

```python
# counts how many pokemon there are for each dual typing combination

df['count'] = 1
df.groupby(['Type 1', 'Type 2']).count()['count']
```

# Working with large amounts of data

```python
# we can load in the dataset in chunks 

for df in pd.read_csv('modified.csv', chunksize = 5): # reads the first 5 rows of data
    print('Chunk DataFrame:')
    print(df)
```

```python
# create new data frame with the same columns
new_df = pd.DataFrame(columns = df.columns)

for df in pd.read_csv('modified.csv', chunksize = 5): 
    
    results = df.groupby(['Type 1']).count() # analyze each chunk of data for results
    new_df = pd.concat([new_df, results]) # add results to new data frame    
```
